```python
df = master_df



# Constants
cost_per_call = 100
budget = 100000
new_sales_reps = 110

# Step 3: Calculate revenue covered by Zederma and Mercitelo
df['Zederma_Coverage'] = df['Zederma TRx'] / df['Topical Analgesic market TRx'].replace(0, 1)  # Avoid division by zero
df['Mercitelo_Coverage'] = df['Mercitelo TRx'] / df['Insomnia Market TRx'].replace(0, 1)  # Avoid division by zero

# Step 4: Calculate amount spent on calling each HCP (cost/call * TRx in each market)
df['Zederma_Cost'] = df['Topical Analgesic market TRx'] * cost_per_call
df['Mercitelo_Cost'] = df['Insomnia Market TRx'] * cost_per_call

# Step 5: Calculate ROI for Zederma and Mercitelo
df['Zederma_ROI'] = df['Zederma TRx'] / df['Zederma_Cost'].replace(0, 1)  # Avoid division by zero
df['Mercitelo_ROI'] = df['Mercitelo TRx'] / df['Mercitelo_Cost'].replace(0, 1)  # Avoid division by zero

# Step 6: Adjust for new sales reps and budget limit
# Simulate adding more sales reps and covering more deciles based on their cost
def calculate_roi_with_budget(df, budget, cost_per_call):
    total_cost = df['Zederma_Cost'].sum() + df['Mercitelo_Cost'].sum()

    if total_cost > budget:
        # Adjust by selecting lower decile (i.e., remove lowest-revenue HCPs to fit within budget)
        df_sorted = df.sort_values(by=['Zederma_Coverage', 'Mercitelo_Coverage'], ascending=False)
        running_cost = 0
        selected_hcps = []
        for index, row in df_sorted.iterrows():
            running_cost += row['Zederma_Cost'] + row['Mercitelo_Cost']
            if running_cost > budget:
                break
            selected_hcps.append(index)
        df_selected = df.loc[selected_hcps]
    else:
        df_selected = df

    total_revenue = df_selected['Zederma TRx'].sum() + df_selected['Mercitelo TRx'].sum()
    total_cost_selected = df_selected['Zederma_Cost'].sum() + df_selected['Mercitelo_Cost'].sum()

    return total_revenue / total_cost_selected

# Initial ROI with budget
initial_roi = calculate_roi_with_budget(df, budget, cost_per_call)

# Adjusted ROI with 110 sales reps
new_budget = new_sales_reps * (df['Topical Analgesic market TRx'].mean() * cost_per_call)
adjusted_roi = calculate_roi_with_budget(df, new_budget, cost_per_call)

# Print results
print(f"Initial ROI: {initial_roi}")
print(f"Adjusted ROI with 110 Sales Reps: {adjusted_roi}")
```

```python

tiers = {
    'Tester': (0, 5, 0.01),
    'Explorer': (5, 10, 0.03),
    'Advocate': (10, float('inf'), 0.05)
}

# Calculate Total Market TRx per Account
master_df['Total Market TRx'] = (
    master_df['Zederma TRx'] +
    master_df['Topical Analgesic market TRx'] +
    master_df['Mercitelo TRx'] +
    master_df['Insomnia Market TRx']
)

# Calculate Product Prescription Percentage
master_df['Zederma Percentage'] = (master_df['Zederma TRx'] / master_df['Total Market TRx']) * 100
master_df['Mercitelo Percentage'] = (master_df['Mercitelo TRx'] / master_df['Total Market TRx']) * 100

# Determine Tier for Each Account
def determine_tier(row):
    max_percentage = max(row['Zederma Percentage'], row['Mercitelo Percentage'])
    for tier, (min_perc, max_perc, discount) in tiers.items():
        if min_perc <= max_percentage < max_perc:
            return tier
    return 'None'

master_df['Tier'] = master_df.apply(determine_tier, axis=1)

# Count of Accounts per Tier
tier_counts = master_df['Tier'].value_counts().reset_index()
tier_counts.columns = ['Tier', 'Count']

# Apply Discounts
def apply_discount(row):
    if row['Tier'] == 'Tester':
        return row['Zederma TRx'] * 0.01
    elif row['Tier'] == 'Explorer':
        return row['Zederma TRx'] * 0.03
    elif row['Tier'] == 'Advocate':
        return row['Zederma TRx'] * 0.05
    return 0

master_df['Discount Applied'] = master_df.apply(apply_discount, axis=1)

# Show results
print(tier_counts)
print(master_df[['Physician ID', 'Total Market TRx', 'Tier', 'Discount Applied']])

```

```python
import pandas as pd

data1 = pd.read_csv('1.csv')
data2 = pd.read_csv('2.csv')

data1.rename(columns={'Zip Code': 'Zip'}, inplace=True)


for col in ['Zederma TRx', 'Topical Analgesic market TRx', 'Mercitelo TRx', 'Insomnia Market TRx']:
    data1[col] = pd.to_numeric(data1[col], errors='coerce').fillna(0)

# Merge the datasets on the 'Zip' column
merged_df = pd.merge(data1, data2, on='Zip', how='inner')


for col in ['Zederma TRx', 'Topical Analgesic market TRx', 'Mercitelo TRx', 'Insomnia Market TRx']:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce').fillna(0)

territory_sales = merged_df.groupby('Territory').agg({
    'Zederma TRx': 'sum',
    'Topical Analgesic market TRx': 'sum',
    'Mercitelo TRx': 'sum',
    'Insomnia Market TRx': 'sum'
}).reset_index()

# weighted sales for each territory
territory_sales['Weighted_Sales'] = (
    0.8 * (territory_sales['Zederma TRx'] + territory_sales['Mercitelo TRx']) +
    0.2 * (territory_sales['Topical Analgesic market TRx'] + territory_sales['Insomnia Market TRx'])
)

#  weighted sales to get the workload index
total_weighted_sales = territory_sales['Weighted_Sales'].sum()
territory_sales['Workload_Index'] = (territory_sales['Weighted_Sales'] / total_weighted_sales) * 60000

# Output the results
print("Workload Index for each Territory:")
print(territory_sales[['Territory', 'Workload_Index']])
```

```python
lower_bound = 800
upper_bound = 1200

balanced_territories = territory_sales[(territory_sales['Workload_Index'] >= lower_bound) &
                                       (territory_sales['Workload_Index'] <= upper_bound)]

above_balanced = territory_sales[territory_sales['Workload_Index'] > upper_bound]
below_balanced = territory_sales[territory_sales['Workload_Index'] < lower_bound]

# Output the results
print(f"Number of Territories within balanced range ({lower_bound}-{upper_bound}): {balanced_territories.shape[0]}")
print(f"Number of Territories above the balanced range: {above_balanced.shape[0]}")
print(f"Number of Territories below the balanced range: {below_balanced.shape[0]}")
```

```python
import matplotlib.pyplot as plt

sorted_territory_sales = territory_sales.sort_values(by='Workload_Index', ascending=False)

plt.figure(figsize=(10, 12))
plt.barh(sorted_territory_sales['Territory'], sorted_territory_sales['Workload_Index'], color='skyblue')
plt.xlabel('Workload Index')
plt.ylabel('Territory')
plt.title('Workload Index for All Territories')
plt.gca().invert_yaxis()
plt.show()
```
